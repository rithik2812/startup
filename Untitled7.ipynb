```python
# ============================================================
# üìå Startup Ecosystem Forecasting with ARIMA & LSTM
#     Using "Startup Growth & Funding Trends" dataset
# ============================================================

# 0Ô∏è‚É£ IMPORTS
import numpy as np
import pandas as pd
import io

import matplotlib.pyplot as plt
from math import sqrt

from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import MinMaxScaler

from statsmodels.tsa.arima.model import ARIMA

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

from google.colab import files

# For reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# ============================================================
# 1Ô∏è‚É£ UPLOAD DATASET (same style as your ATM snippet)
# ============================================================

print("üì§ Please upload the Startup dataset CSV file (e.g. startup_data.csv or startup_growth_data.csv)")
uploaded = files.upload()   # Shows "Choose Files" button in Colab

csv_name = list(uploaded.keys())[0]
print("üìÅ Using file:", csv_name)

# ---- Robust CSV loading: try utf-8, fall back to latin-1 ----
raw_bytes = uploaded[csv_name]

try:
    df = pd.read_csv(io.BytesIO(raw_bytes), low_memory=False)
    print("‚úÖ Loaded CSV with default UTF-8 decoding.")
except UnicodeDecodeError:
    print("‚ö†Ô∏è UTF-8 decoding failed. Trying 'latin-1' encoding...")
    df = pd.read_csv(io.BytesIO(raw_bytes), encoding="latin-1", low_memory=False)
    print("‚úÖ Loaded CSV with 'latin-1' decoding.")

print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print(df.head())

# ============================================================
# 2Ô∏è‚É£ BASIC CLEANING & COLUMN STANDARDIZATION
# ============================================================

# Strip whitespace from column names
df.columns = [c.strip() for c in df.columns]

print("\nüßæ Cleaned column names:", df.columns.tolist())

# Adjust these if your column names differ slightly
COL_YEAR       = "Year Founded"
COL_NAME       = "Startup Name"
COL_FUNDING    = "Funding Amount (M USD)"
COL_VALUATION  = "Valuation (M USD)"
COL_REVENUE    = "Revenue (M USD)"
COL_PROFITABLE = "Profitable"

# Sanity check
required_cols = [COL_YEAR, COL_NAME, COL_FUNDING, COL_VALUATION, COL_REVENUE, COL_PROFITABLE]
for col in required_cols:
    if col not in df.columns:
        raise ValueError(f"‚ùå Required column '{col}' not found. Found: {df.columns.tolist()}")

# Drop rows with missing key fields (conservative)
df = df.dropna(subset=[COL_YEAR, COL_FUNDING, COL_VALUATION, COL_REVENUE, COL_PROFITABLE])
df[COL_YEAR] = df[COL_YEAR].astype(int)

print("\n‚úÖ After cleaning, shape:", df.shape)

# ============================================================
# 3Ô∏è‚É£ BUILD YEARLY TIME-SERIES AGGREGATES
# ============================================================

df_year = (
    df.groupby(COL_YEAR)
      .agg(
          n_startups   = (COL_NAME, "count"),
          total_funding= (COL_FUNDING, "sum"),
          avg_valuation= (COL_VALUATION, "mean"),
          avg_revenue  = (COL_REVENUE, "mean"),
          profitability_rate = (COL_PROFITABLE, "mean"),  # fraction of profitable
          unicorn_count      = (COL_VALUATION, lambda x: (x >= 1000).sum())
      )
      .reset_index()
      .rename(columns={COL_YEAR: "year"})
      .sort_values("year")
)

# Just in case, ensure no NaNs in total_funding (forward fill + 0 fallback)
df_year["total_funding"] = df_year["total_funding"].fillna(method="ffill").fillna(0.0)

print("\nüìà Yearly aggregated data (head):")
print(df_year.head())
print("\nüìà Yearly aggregated data (tail):")
print(df_year.tail())

# ============================================================
# 4Ô∏è‚É£ VISUALIZE KEY TRENDS
# ============================================================

plt.figure(figsize=(10, 5))
plt.plot(df_year["year"], df_year["total_funding"], marker="o")
plt.title("Total Funding per Year (M USD)")
plt.xlabel("Year")
plt.ylabel("Total Funding (M USD)")
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(df_year["year"], df_year["avg_valuation"], marker="o")
plt.title("Average Valuation per Year (M USD)")
plt.xlabel("Year")
plt.ylabel("Average Valuation (M USD)")
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(df_year["year"], df_year["unicorn_count"], marker="o")
plt.title("Number of Unicorns per Year")
plt.xlabel("Year")
plt.ylabel("Unicorn Count (Valuation ‚â• 1000 M USD)")
plt.grid(True)
plt.show()

# ============================================================
# 5Ô∏è‚É£ ARIMA FORECASTING ‚Äì TOTAL FUNDING PER YEAR (FIXED)
# ============================================================

ts_funding = df_year.set_index("year")["total_funding"].astype(float).sort_index()

print("\nüìä Time series length:", len(ts_funding))
print(ts_funding.head())

# Train-test split (80/20 by time)
n = len(ts_funding)
train_size = int(n * 0.8)
train_ts = ts_funding.iloc[:train_size]
test_ts  = ts_funding.iloc[train_size:]

print(f"\nüîπ Total points: {n}, Train: {len(train_ts)}, Test: {len(test_ts)}")

# ---- Fit ARIMA on plain values (avoid index issues) ----
arima_order = (1, 1, 1)  # can tune later

print("\n‚è≥ Fitting ARIMA model...")
arima_model = ARIMA(train_ts.values, order=arima_order)
arima_res = arima_model.fit()
print("‚úÖ ARIMA model fitted.")
print(arima_res.summary())

# ---- Forecast on test horizon ----
arima_forecast_values = arima_res.forecast(steps=len(test_ts))
arima_forecast = pd.Series(arima_forecast_values, index=test_ts.index)

print("\n‚ùì NaNs in test or forecast?")
print("   test_ts NaNs  :", int(test_ts.isna().sum()))
print("   forecast NaNs :", int(arima_forecast.isna().sum()))

# Mask NaNs if any
mask = (~test_ts.isna()) & (~arima_forecast.isna())
clean_test = test_ts[mask]
clean_forecast = arima_forecast[mask]

if len(clean_test) == 0:
    raise ValueError("All test or forecast values are NaN ‚Äì ARIMA failed. Try different (p,d,q).")

# ---- Evaluation ----
arima_rmse = sqrt(mean_squared_error(clean_test, clean_forecast))
arima_mae  = mean_absolute_error(clean_test, clean_forecast)

print(f"\nüìä ARIMA Performance on Test Set (Total Funding):")
print(f"   RMSE: {arima_rmse:.2f}")
print(f"   MAE : {arima_mae:.2f}")

# ---- Plot Train/Test/Forecast ----
plt.figure(figsize=(10, 5))
plt.plot(train_ts.index, train_ts.values, label="Train")
plt.plot(test_ts.index, test_ts.values, label="Test", marker="o")
plt.plot(arima_forecast.index, arima_forecast.values, label="ARIMA Forecast", marker="x")
plt.title("ARIMA Forecast ‚Äì Total Funding per Year")
plt.xlabel("Year")
plt.ylabel("Total Funding (M USD)")
plt.legend()
plt.grid(True)
plt.show()

# ---- Future forecast (next 5 years) ----
future_steps = 5
future_values_all = arima_res.forecast(steps=len(test_ts) + future_steps)
future_values_only = future_values_all[-future_steps:]

last_year = ts_funding.index.max()
future_years_arima = list(range(last_year + 1, last_year + 1 + future_steps))

arima_future_only = pd.Series(future_values_only, index=future_years_arima)

print("\nüîÆ ARIMA Future Forecast ‚Äì Total Funding (Next 5 Years):")
print(arima_future_only)

plt.figure(figsize=(10, 5))
plt.plot(ts_funding.index, ts_funding.values, label="Historical")
plt.plot(arima_future_only.index, arima_future_only.values, label="ARIMA Future Forecast", marker="o")
plt.title("ARIMA Future Forecast ‚Äì Total Funding per Year")
plt.xlabel("Year")
plt.ylabel("Total Funding (M USD)")
plt.legend()
plt.grid(True)
plt.show()

# ============================================================
# 6Ô∏è‚É£ LSTM FORECASTING ‚Äì TOTAL FUNDING PER YEAR
# ============================================================

# Convert series to numpy array and scale
values = ts_funding.values.reshape(-1, 1)

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_values = scaler.fit_transform(values)

def create_sequences(data, lookback=3):
    """
    Turn a 1D sequence into overlapping sequences of length `lookback`
    to predict the next point.
    """
    X, y = [], []
    for i in range(len(data) - lookback):
        X.append(data[i:i + lookback])
        y.append(data[i + lookback])
    return np.array(X), np.array(y)

lookback = 3

# Split scaled_values into train/test (same 80/20 split, but keep overlap for sequences)
n_scaled = len(scaled_values)
train_size_scaled = int(n_scaled * 0.8)

train_scaled = scaled_values[:train_size_scaled]
test_scaled  = scaled_values[train_size_scaled - lookback:]  # overlap

X_train, y_train = create_sequences(train_scaled, lookback)
X_test,  y_test  = create_sequences(test_scaled, lookback)

print("\nüì¶ LSTM sequence shapes:")
print("   X_train:", X_train.shape)
print("   y_train:", y_train.shape)
print("   X_test :", X_test.shape)
print("   y_test :", y_test.shape)

# ---- Build LSTM model ----
model = Sequential([
    LSTM(32, input_shape=(lookback, 1)),
    Dense(1)
])

model.compile(optimizer="adam", loss="mse")
model.summary()

print("\n‚è≥ Training LSTM model...")
history = model.fit(
    X_train, y_train,
    epochs=100,          # reduce to ~40 if too slow
    batch_size=8,
    validation_split=0.1,
    verbose=1
)
print("‚úÖ LSTM training complete.")

# ---- Predict on test data ----
y_pred_scaled = model.predict(X_test)

# Inverse transform to original scale
y_test_inv = scaler.inverse_transform(y_test)
y_pred_inv = scaler.inverse_transform(y_pred_scaled)

# ---- Align years for LSTM test predictions ----
test_years = ts_funding.index[train_size:]         # test years
test_years_for_seq = test_years[lookback:]        # lose first 'lookback' due to sequences

min_len = min(len(test_years_for_seq), len(y_test_inv))
test_years_for_seq = test_years_for_seq[:min_len]
y_test_inv = y_test_inv[:min_len]
y_pred_inv = y_pred_inv[:min_len]

# ---- LSTM metrics ----
lstm_rmse = sqrt(mean_squared_error(y_test_inv, y_pred_inv))
lstm_mae  = mean_absolute_error(y_test_inv, y_pred_inv)

print(f"\nüìä LSTM Performance on Test Set (Total Funding):")
print(f"   RMSE: {lstm_rmse:.2f}")
print(f"   MAE : {lstm_mae:.2f}")

# ---- Plot Actual vs LSTM Predicted ----
plt.figure(figsize=(10, 5))
plt.plot(test_years_for_seq, y_test_inv.flatten(), label="Actual (Test)", marker="o")
plt.plot(test_years_for_seq, y_pred_inv.flatten(), label="LSTM Predicted", marker="x")
plt.title("LSTM Forecast ‚Äì Total Funding per Year (Test Period)")
plt.xlabel("Year")
plt.ylabel("Total Funding (M USD)")
plt.legend()
plt.grid(True)
plt.show()

# ============================================================
# 7Ô∏è‚É£ LSTM FUTURE FORECAST (NEXT 5 YEARS)
# ============================================================

future_steps = 5

# Start from last `lookback` scaled points
last_sequence = scaled_values[-lookback:]
current_seq = last_sequence.reshape(1, lookback, 1)

future_preds_scaled = []

for _ in range(future_steps):
    next_pred_scaled = model.predict(current_seq)
    future_preds_scaled.append(next_pred_scaled[0, 0])
    # slide window
    current_seq = np.append(current_seq[:, 1:, :], [[next_pred_scaled]], axis=1)

future_preds_scaled = np.array(future_preds_scaled).reshape(-1, 1)
future_preds = scaler.inverse_transform(future_preds_scaled)

last_year = ts_funding.index.max()
future_years_lstm = list(range(last_year + 1, last_year + 1 + future_steps))

print("\nüîÆ LSTM Future Forecast ‚Äì Total Funding (Next 5 Years):")
for y, val in zip(future_years_lstm, future_preds.flatten()):
    print(f"   {y}: {val:.2f} M USD")

plt.figure(figsize=(10, 5))
plt.plot(ts_funding.index, ts_funding.values, label="Historical")
plt.plot(future_years_lstm, future_preds.flatten(), label="LSTM Future Forecast", marker="o")
plt.title("LSTM Future Forecast ‚Äì Total Funding per Year")
plt.xlabel("Year")
plt.ylabel("Total Funding (M USD)")
plt.legend()
plt.grid(True)
plt.show()

# ============================================================
# 8Ô∏è‚É£ FINAL COMPARISON SUMMARY
# ============================================================

print("\n==============================================")
print("üìä ARIMA vs LSTM ‚Äì Total Funding per Year")
print("==============================================")
print(f"ARIMA RMSE: {arima_rmse:.2f}, MAE: {arima_mae:.2f}")
print(f"LSTM  RMSE: {lstm_rmse:.2f}, MAE: {lstm_mae:.2f}")
print("==============================================")
```

```python
# ============================================================
# ‚úÖ Percentage Error Metrics for ARIMA & LSTM
# ============================================================

# ---- 1Ô∏è‚É£ Get the average actual value in the test set ----
mean_actual = clean_test.mean()

print("\nüìå Average Actual Funding in Test Period:", mean_actual)

# ---- 2Ô∏è‚É£ Convert ARIMA errors to percentage ----
arima_rmse_pct = (arima_rmse / mean_actual) * 100
arima_mae_pct  = (arima_mae  / mean_actual) * 100

# ---- 3Ô∏è‚É£ Convert LSTM errors to percentage ----
lstm_rmse_pct = (lstm_rmse / mean_actual) * 100
lstm_mae_pct  = (lstm_mae  / mean_actual) * 100

# ---- 4Ô∏è‚É£ Print results ----
print("\n==============================================")
print("üìä Percentage Error Metrics")
print("==============================================")
print(f"ARIMA RMSE%: {arima_rmse_pct:.2f}%")
print(f"ARIMA MAE% : {arima_mae_pct:.2f}%")
print("----------------------------------------------")
print(f"LSTM RMSE% : {lstm_rmse_pct:.2f}%")
print(f"LSTM MAE%  : {lstm_mae_pct:.2f}%")
print("----------------------------------------------")

# ---- 5Ô∏è‚É£ Optional: LSTM improvement vs ARIMA ----
improvement_rmse_pct = ((arima_rmse - lstm_rmse) / arima_rmse) * 100
improvement_mae_pct  = ((arima_mae  - lstm_mae)  / arima_mae ) * 100

print(f"‚úÖ LSTM RMSE Improvement: {improvement_rmse_pct:.2f}%")
print(f"‚úÖ LSTM MAE Improvement : {improvement_mae_pct:.2f}%")
print("==============================================")
```
